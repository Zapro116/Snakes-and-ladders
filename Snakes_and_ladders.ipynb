```python
import random
import sys

snakes={
    17 : 7,
    54 : 34,
    62 : 19,
    98 : 79
}

ladders={
    3 : 38,
    24 : 33,
    42 : 93,
    72 : 84
}

def welcome_msg():
    print("#"*90)
    print("*"*90)
    print("#"*90)
    msg="""
    Welcome to snakes and ladders game
    Made by Hariom Chaurasia
    Roll 1806116
    So let's begin our game!!!
    """
    print(msg)
    print("#"*90)
    print("*"*90)
    print("#"*90)


def instructions_and_rules():
    print("*"*90)
    inst = """
    Enter "roll" to roll the dice
    For manual mode you may type any number between 1 to 20 (NOT GREATER THAN 20)
    Enter "quit" to quit the game immidiately
    Enter "help" for this dialogue box to appear
    """
    print(inst)
    print("*"*90)

def dice_roll(player):
    x = random.randint(1,6)
    print(f"    {player} got "+ str(x) +" !!!")
    return x


def player_aboard():
    welcome_msg()
    print("    Before we start, let's know your names!!!")
    p1=""
    b1=0
    b2=0
    p2=""
    while(not p1):
        if(b1 == 0):
            b1=1
        else:
            print("YOU CAN'T LEAVE NAME EMPTY!!!")
        p1 = input("    Enter your name player1 : ")
    b2=0
    while(not p2):
        if(b2 == 0):
            b2 = 1
        else:
            print("YOU CAN'T LEAVE NAME EMPTY!!!")
        p2 = input("    Enter your name player2 : ")
    print("   "+"*"*70)
    print(f"    Match will be played between {p1} and {p2}")
    print("   "+"*"*70)
    instructions_and_rules()
    return p1,p2

def climb_ladder():
    print("Congrats!!!\nYou found a ladder\nClimbing the ladder ###")

def snake_bite():
    print("Ssssshhhhh\nA snake bit you ~~~")
    
def player_move(player,current_value,dice_value):
    old_value=current_value
    current_value+=dice_value
    
    if(current_value>100):
        current_value=old_value
        print(f"You need {100-old_value} to win the game\nkeep trying, you are really close!!!")
        return old_value
    
    if current_value in snakes:
        final_value=snakes.get(current_value)
        snake_bite()
    elif current_value in ladders:
        final_value=ladders.get(current_value)
        climb_ladder()
    else:
        final_value=current_value
    
    return final_value

def check_win(player_name, position):
    if position == 100:
        print("#"*90)
        print("*"*90)
        print("#"*90)
        print("Finally!!! Our winner is here\n\n" + player_name + " won the game.")
        print("\nCongratulations " + player_name)
        print("#"*90)
        print("*"*90)
        print("#"*90)
        return sys.exit(1)

def input_check(player,player_pos,inp):
    if(inp=="roll"):
        return player_pos + dice_roll(player)
    elif inp.isnumeric():
        inp1 = int(inp)
        if (inp1 >= 1 and inp1 <= 20):
            return player_move(player,player_pos,inp1)
        else:
            print("\nCan't you read English!!!\nEnter a number between 1 and 20 for manual mode")
            inp1=input(f"Enter your choice {player} : ")
            return input_check(player,player_pos,inp1)
    elif(inp=="help"):
        instructions_and_rules()
        inp=input(f"Enter your choice {player} : ")
        return input_check(player,player_pos,inp)
    elif(inp=="quit"):
        sys.exit(1)
    else:
        print("Invalid input!!!\nTry again")
        inp=input(f"Enter your choice {player} : ")
        return input_check(player,player_pos,inp)

        

def main():
    p1,p2 = player_aboard()
    p1_pos=0
    p2_pos=0
    while True:
        inp = input(f"Enter your choice {p1} : ")
        p1_pos=input_check(p1,p1_pos,inp)
        print(f"   {p1} is at position "+str(p1_pos))
        print("*"*70)
        check_win(p1,p1_pos)
        inp = input(f"Enter your choice {p2} : ")
        p2_pos=input_check(p2,p2_pos,inp)
        print(f"   {p2} is at position "+str(p2_pos))
        print("*"*70)
        check_win(p2,p2_pos)
   

main()
```
